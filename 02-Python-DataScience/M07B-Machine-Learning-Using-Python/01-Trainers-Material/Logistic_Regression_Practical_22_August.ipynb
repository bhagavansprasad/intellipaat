### **Check for the outliers**

```python
data.describe()   #Statistical Summary of the Data
```

**Observation:** It seems that columns trestbps, chol, thalach, oldpeak has some outlier which we can verify by making boxplots.

As example we will the outlier removal process using `chol` column.

```python
q1 = np.percentile(data['chol'], 25)
q3 = np.percentile(data['chol'], 75)
```

```python
print(q1, q3)
```

```python
iqr = q3 - q1
```

```python
LB = q1 - 1.5 * iqr
UB = q3 + 1.5 * iqr
```

```python
print(LB, UB)
```

```python
(data['chol'] >= LB) & (data['chol'] <= UB)
```

```python
data = data[(data['chol'] >= LB) & (data['chol'] <= UB)]
```
